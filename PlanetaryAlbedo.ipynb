```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

%matplotlib inline
```

```python
# Define alpha's
tau = np.arange(0,50,0.5)
gamma = 1
alpha_prime = gamma*tau/(1 + gamma*tau)
alpha_a = 1 - np.exp(-tau*sqrt(2))
alpha_g1 = 0.3
alpha_g2 = 0.8
```

```python
# Define planetary albedo
def net_albedo(x):
    return ((x*(1 - alpha_prime)*(1 - alpha_a))/(1 - alpha_prime*x)) + alpha_a
```

```python
# Calculate net albedos for different alpha_g's
alpha1 = net_albedo(alpha_g1)
alpha2 = net_albedo(alpha_g2)
```

```python
plt.plot(tau, alpha1, 'b-', label = 'alpha_g = 0.3')
plt.plot(tau, alpha2, 'r-', label = 'alpha_g = 0.8')
plt.legend(loc = 'lower right')
plt.xlabel('tau_inf')
plt.ylabel('Net Planetary Albedo')
plt.grid()
plt.show()
```
